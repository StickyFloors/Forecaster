8. Daily Max

```python

daily_max = requests.get("https://api.open-meteo.com/v1/forecast?latitude=51.46&longitude=0.44000006&daily=temperature_2m_max&hourly=temperature_2m,apparent_temperature&forecast_days=3")
daily_max_data = daily_max.json()
print(daily_max_data)
```

```python
three_day_high=dict(zip(daily_max_data['daily']['time'],daily_max_data['daily']['temperature_2m_max']))
print(three_day_high)

for key,value in three_day_high.items():
    print(f"The high for {key} is {value}.")
print(f"The hottest day will be {max(three_day_high,key=three_day_high.get)}.")
```
